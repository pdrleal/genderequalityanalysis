```python
import pandas as pd
import numpy as np
```

```python
def map_yearcolumn(x):
    return x[0:4];
```

```python
df = pd.read_excel('P_Data_Extract_From_Gender_Statistics.xlsx')
```

```python
df.replace('..',np.NaN,inplace=True)
```

```python
df.drop("Series Code",axis=1,inplace=True)
```

```python
years_columns= df.iloc[:,3:].columns.tolist()
info_columns=df.iloc[:,0:3].columns.tolist()
```

```python
info_columns
```

```python
years_columns
```

```python
df.columns=info_columns + list(map(map_yearcolumn,years_columns))
years_columns= df.iloc[:,3:].columns.tolist()
```

```python
df[years_columns].isna().sum().plot()
```

```python
range_of_years=list(map(str,np.arange(2000,2021)))
df[range_of_years].isna().sum().plot()
```

```python
df
```

```python
df = df[info_columns +range_of_years]
years_columns= df.iloc[:,3:].columns.tolist()
```

```python
df.head(2)
```

```python
df_2 = df[info_columns+range_of_years].copy()
```

```python
df_2
```

```python
df_long = pd.melt(df_2, id_vars=['Series Name', 'Country Name','Country Code'], var_name='Year', value_name='Value')
# pivot the dataframe into the desired format
df_pivot = pd.pivot_table(df_long, index=['Country Name','Country Code', 'Year'], columns='Series Name', values='Value')
# reset the index to make the columns 'country_name' and 'year' regular columns
df_pivot = df_pivot.reset_index()
df_pivot
```

```python
df_pivot.to_excel('../dash/assets/Data_Gender_Statistics.xlsx',index=False)
```
